# FEATURE SELECTION

UNIVARIATE FEATURES


```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

# READ DATA

```python
data = pd.read_csv("./train.csv")
```

```python
data.head(n=5)
```

```python
X = data.iloc[:,0:20]
y = data.iloc[:,-1]
```

```python
print(X.shape,y.shape)
```

# _CHOOSING BEST FEATURES_

# 1. USING CHI SQUARE

```python
best_features = SelectKBest(score_func = chi2,k=10)
```

```python
fit = best_features.fit(X,y)
```

```python
fit.scores_
```

```python
scores =pd.DataFrame(fit.scores_)
```

```python
scores_col = pd.DataFrame(X.columns)
```

```python
scores = pd.concat([scores,scores_col],axis=1)
```

```python
scores.columns = ['score','feature']
```

```python
scores=scores.sort_values(by='score',ascending=False)
```

```python
scores
```

#  2.USING IMPORTANT FEATURES(RANDOM FOREST)

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
model = RandomForestClassifier()
model.fit(X,y)
```

```python
model.feature_importances_
```

```python
important_features = pd.DataFrame(model.feature_importances_,index=X.columns)
```

```python
important_features.columns = ['Importance']
```

```python
important_features
```

```python
important_features = important_features.sort_values(by='Importance',ascending=False)
```

```python
important_features
```

```python
plt.style.use("seaborn")
plt.figure(figsize=(20,10))
plt.bar(important_features.index,important_features['Importance'])
plt.show()
```

# CHECK THE MODEL PERFORMANCE

```python
from sklearn.model_selection import cross_val_score
```

```python
rfc = RandomForestClassifier()
scores_new = cross_val_score(rfc , X , y ,cv = 10)
```

```python
scores_new
```

```python
#this is accuracy
scores_new.mean()
```

```python
#convert previous scores into a numpy array
scores = scores.values
```

```python
#take most 10 features
scores = scores[0:10]
```

```python
score_compare=[]
for score in scores:
    score_compare.append(score[1])
```

```python
X[score_compare]
```

```python
print((cross_val_score(rfc,X[score_compare],y,cv=10)).mean())
#we got accuracy of 90%
```
