```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
import seaborn as sns
```

# Batch Gradient Descent

```python
X,y = make_regression(n_samples=10000,n_features=20,n_informative=20,noise=0)
```

```python
X.shape,y.shape
plt.scatter(X[:,5],y)
plt.show()
```

```python
Xo = np.ones((X.shape[0],1))
X = np.hstack((Xo,X))
X.shape,y.shape
```

# Batch GD

```python
def hypothesis(X,theta):
    return np.dot(X,theta)
def error(X,y,theta):
    e = 0.0
    return (np.sum((hypothesis(X,theta)-y)**2))/X.shape[0]
def gradient(X,y,theta):
    grad = np.dot(X.T,hypothesis(X,theta)-y)
    return grad/X.shape[0]
def gradient_descent(X,y,learn_rate=0.01,epochs=1000):
    error_list = []
    theta = np.zeros((X.shape[1]),)
    for i in range(epochs):
        e = error(X,y,theta)
        error_list.append(e)
        theta  = theta - learn_rate*gradient(X,y,theta)
    return theta,error_list
```

```python
theta,error_list = gradient_descent(X,y)
```

# Smooth Trajectory

```python
plt.plot(error_list)
plt.show()
```

#  Mini Batch Gradient Descent

```python
#shuffle the data
#first make a single matrix from X,y
data = np.hstack((X,y.reshape(-1,1)))
#now shuffle
np.random.shuffle(data)
```

```python
def mini_gradient_descent(X,y,learn_rate=0.01,epochs=10,batch_size = 500):
    m = data.shape[0]
    total_batches = m//batch_size
    
    error_list = []
    theta = np.zeros((X.shape[1]),)
    for j in range(epochs):
        for i in range(total_batches):
            batch_data = data[i*batch_size:(i+1)*batch_size,:]
            x_batch = batch_data[:,:-1]
            y_batch = batch_data[:,-1]
            batch_gradient = gradient(x_batch,y_batch,theta)
            theta = theta - learn_rate*batch_gradient
            e = error(x_batch,y_batch,theta)
            error_list.append(e)
    return theta,error_list
```

# Noisy gradient plot Trajectory

```python
theta,error_list = mini_gradient_descent(X,y)
plt.plot(np.arange(len(error_list)),error_list)
plt.show()
```

- if batch_size = 1 then schotastic GD
- if batch_size = total number of examples then Batch GD

# Advantages of Mini Batch
- Faster Convergence
- Noisy Updates help to avoid local minima
