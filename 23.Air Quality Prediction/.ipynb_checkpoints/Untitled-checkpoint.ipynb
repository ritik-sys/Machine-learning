```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./Train.csv")
```

```python
dataset.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
dataset = dataset.values
```

```python
X,y = dataset[:,0:4],dataset[:,5]
```

```python
X.shape,y.shape
```

```python
plt.figure(figsize=(5,10))
plt.subplot(5,1,1)
plt.scatter(X[:,0],y)
plt.subplot(5,1,2)
plt.scatter(X[:,1],y)
plt.subplot(5,1,3)
plt.scatter(X[:,2],y)
plt.subplot(5,1,4)
plt.scatter(X[:,3],y)
plt.plot
```

```python
#normalize data
meu = np.mean(X,axis=0)
std = np.std(X,axis=0)
X = (X-meu)/std
```

# PREDICTIONS


```python
model = LinearRegression()
model.fit(X,y)
```

```python
coeff=list(model.coef_)
intersept = model.intercept_
```

```python
theta = []
theta.append(intersept)
for i in range(len(coeff)):
    theta.append(coeff[i])
```

```python
theta = np.array(theta)
```
