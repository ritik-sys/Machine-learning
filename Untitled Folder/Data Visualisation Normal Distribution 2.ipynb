```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#comparision of spread for different values of sigma
x = np.random.randn(100)
x = 10 + 5*x
y = np.zeros(100)
plt.scatter(x,y)
plt.show()
plt.hist(x)
plt.show()
```
