# Multivariate Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#this is standard form with ux and uy both zero and covariance is a identity matrix
u1 = np.array([0,0])
sig1 = np.array([[1,0],[0,1]])
dist1 = np.random.multivariate_normal(u1,sig1,1000)
plt.scatter(dist1[:,0],dist1[:,1])
plt.show()
```

```python
#Lets change ux and uy and compare with previous one
u1 = np.array([0,0])
sig1 = np.array([[1,0],[0,1]])
u2 = np.array([10,5])
sig2 = np.array([[1,0],[0,1]])
dist1 = np.random.multivariate_normal(u1,sig1,1000)
dist2 = np.random.multivariate_normal(u2,sig2,1000)
plt.scatter(dist1[:,0],dist1[:,1])
plt.scatter(dist2[:,0],dist1[:,1])
plt.show()
```

```python
#Lets change covariance matrix at 45 degree
u1 = np.array([0,0])
sig1 = np.array([[1,0.5],[0.5,1]])
dist1 = np.random.multivariate_normal(u1,sig1,1000)
plt.scatter(dist1[:,0],dist1[:,1])
u2 = np.array([0,0])
sig2 = np.array([[1,-0.5],[-0.5,1]])
dist2 = np.random.multivariate_normal(u2,sig2,1000)
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
