```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples = 500 , n_features = 1 , n_informative = 1 , noise = 2)
```

```python
plt.style.use("seaborn")
y = y.reshape(-1,1)
print(X.shape,y.shape)
plt.scatter(X,y)
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
print(X.shape,y.shape)
print(X[:5,:])
```

```python
def hypothesis(X,theta):
    return np.dot(X,theta)
def closed_form_regression(X,y):
    y = np.mat(y)
    first_part = np.dot(X.T,X)
    second_part = np.dot(X.T,y)
    theta = np.linalg.pinv(first_part)*second_part
    return theta
```

```python
theta = closed_form_regression(X,y)
theta
```

```python
#no need to consider X[0] bcz it has ones
req=X[:,1:]
plt.scatter(req,y)
plt.plot(req,hypothesis(X,theta),color="red")
plt.show()
```
