# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
#make a object of Gaussian naive bayse class
gnb = GaussianNB()
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
X,Y=make_classification(n_samples=300,n_features=2,n_informative=2,n_redundant=0,n_classes=2,random_state=4)
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
#continous values features
print(X[0])
```

```python
#fit gaussian model
gnb.fit(X,Y)
```

```python
#this will tell how accurate our model is
print(gnb.score(X,Y))
```

```python
#lets see the predicted values of y
y_pred = gnb.predict(X)
print(y_pred)
```

```python
#original y
Y
```

```python
#lets compare predicted and original y
import numpy as np
print(np.sum(y_pred == Y))
```

```python
#out of 300 , 265 values are correctly mathced
```

# CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
```

```python
cnf_matrix = confusion_matrix(Y,y_pred)
```

```python
cnf_matrix
```

# VISUALIZE CONFUSION MATRIX

```python
import seaborn as sns
sns.heatmap(cnf_matrix/np.sum(cnf_matrix), annot=True, fmt='.2%', cmap='Blues')
```
