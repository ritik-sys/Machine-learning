```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./movie_metadata.csv")
```

```python
df.columns
```

```python
columns = df.get("movie_title")
columns
```

```python
output={}
titles=list(columns)
titles
```

```python
for title in titles:
    length=len(title)
    if length in output.keys():
        output[length]+=1
    else:
        output[length]=1
```

```python
plt.scatter(output.keys(),output.values())
```
