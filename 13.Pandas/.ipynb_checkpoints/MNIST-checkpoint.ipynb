```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load csv
df = pd.read_csv("..datasets/MNIST-2/mnist_train.csv")
```

```python
# convert into numpy array and reshuffle this
data_array = df.values
np.random.shuffle(data_array)
```

```python
# split into X and Y
X = data_array[:,1:]
Y = data_array[:,0]
```

```python
#function to plot one image
def plotImage(X,Y,i):
    plt.imshow(X[i].reshape(28,28),cmap="gray")
    plt.title('Label '+str(Y[i]))
```

```python
#lets plot one image
plotImage(X,Y,0)
```

```python
#split our Dataset into Trainning and Testing part
print(X.shape)
42000*.8
```

```python
plt.figure(figsize=[10,10])
for i in range(25):
    plt.subplot(5,5,i+1)
    plotImage(X,Y,i)
    plt.axis("off")
```
