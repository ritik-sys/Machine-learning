# Scatter Plot


```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

```python
prices = np.array([1,2,3,4])**2
print(prices)
plt.plot(prices)
plt.show()

# if we are not giving both x and y then x will automatically become 
# indices of array
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

```python
#Lets Label the abpove graph
plt.scatter(x,y1,label="Apple",marker="*")
plt.scatter(x,y2,label="Kiwi",marker="o")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price vs time")
plt.legend()
plt.show()
```

```python
# Adjust size of Plot
plt.figure(figsize=(5,5))
plt.scatter(x,y1,label="Apple",marker="*")
plt.scatter(x,y2,label="Kiwi",marker="o")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price vs time")
plt.legend()
plt.show()
```
