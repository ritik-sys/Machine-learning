# Histogram


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 100 points of standard normal distribution(-1,1)
xsn  = np.random.randn(100)
print(xsn)
sigma = 5
mean = 60
print('------------------')
#100 points with sd=5 and mean = 60
x1 = xsn*sigma + mean
np.round(x1,2)
x2 = np.round((xsn*8 + 40),2)
```

```python
plt.style.use('seaborn')
plt.hist(x1,alpha=0.9)
plt.hist(x2,alpha=0.8)
plt.xlabel('Range')
plt.ylabel('Frequency')
plt.show()
```
