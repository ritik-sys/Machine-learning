# Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("default")
```

```python
subjects=["Maths","english","Punjabi","hindi"]
weightage = [20,10,22,2]
plt.pie(weightage,labels=subjects)
plt.show()
```

```python
#explode separates the part of pie
subjects=["Maths","english","Punjabi","hindi"]
weightage = [20,10,22,2]
plt.pie(weightage,labels=subjects,explode=[0,0,0.1,0],autopct="%1.1f%%")
plt.show()
```
