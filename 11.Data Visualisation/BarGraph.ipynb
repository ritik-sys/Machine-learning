# Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
```

```python
plt.bar?
```

```python
plt.bar([0,1,2],[10,20,5]) #current year
plt.bar([0,1,2],[9,13,3]) #next year
plt.show()
```

```python
x_coordinates=np.array([0,1,2])
plt.bar(x_coordinates,[10,20,5],width=0.5,label='cuurent year') #current year
plt.bar(x_coordinates+0.5,[9,13,3],width=0.5,label="next year") #next year
plt.legend()
plt.show()
```

```python
#Tick label
x_coordinates=np.array([0,1,2])
plt.bar(x_coordinates,[10,20,5],width=0.5,label='cuurent year',tick_label=["gold","silver","platinum"]) #current year
plt.bar(x_coordinates+0.5,[9,13,3],width=0.4,label="next year") #next year
plt.legend()
plt.title("Metal price comparision")
plt.show()
```
