# Creating Recommendation engine

```python
df
movie_matrix=pd.pivot_table(df, index = 'user id', values = 'rating',columns='movie_title')
```

```python
rating.sort_values('no of ratings',ascending=False)
```

```python
star_wars_user_rating=movie_matrix['Star Wars (1977)']
```

```python
# how related star wars is with other movies
similar_to_star_wars=movie_matrix.corrwith(star_wars_user_rating)
similar_to_star_wars
```

```python
corr_star_wars=pd.DataFrame(similar_to_star_wars,columns=['correlation'])
```

```python
corr_star_wars.dropna(inplace=True)
```

```python
corr_star_wars
```

```python
corr_star_wars.sort_values('correlation',ascending=False)
```

```python
corr_star_wars=corr_star_wars.join(rating['no of ratings'])
```

```python
corr_star_wars
```

```python
corr_star_wars[corr_star_wars['no of ratings']>100].sort_values(by='correlation',ascending=False)
```

# Predict Function


```python
def predict_movie(movie_name):
    movie_name_user_rating=movie_matrix[movie_name]
    similar_to_movie_name=movie_matrix.corrwith(movie_name_user_rating)
    
    corr_movie_name=pd.DataFrame(similar_to_movie_name,columns=['correlation'])
    corr_movie_name.dropna(inplace=True)
    
    corr_movie_name=corr_movie_name.join(rating['no of ratings'])
    predictions=corr_movie_name[corr_movie_name['no of ratings']>100].sort_values(by='correlation',ascending=False)
    
    return predictions
```

```python
prediction=predict_movie('Titanic (1997)')
```
