```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=1000,noise=10,n_features=2,random_state=1,n_informative=2)
```

```python
X.shape,y.shape
```

### Here first plot is not useful because no linear relation is there
### So y = f(x2)

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)

plt.subplot(1,2,2)
plt.scatter(X[:,1],y)
```

# 3D plot
- if we want to plot interactive plot in jupyter then use _%matplotlib notebook_

```python
%matplotlib notebook
fig = plt.figure()
axes = plt.axes(projection="3d")

axes.scatter3D(X[:,0],X[:,1],y,color="red")
plt.title("3D scatter")
plt.show()
```

# Predictions using sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
#theta1 and theta2
model.coef_
```

```python
#theta0 
model.intercept_
```

```python
#predicted vs original y
print(model.predict([X[0]]))
print(y[0])
```

```python
print(model.predict([X[0],X[2]]))
print(y[0],y[2])
```

# Score

```python
model.score(X,y)
```
