```python
import numpy as np
import matplotlib.pyplot as plt
```

## GRADIENT DESCENT 

```python
x = np.arange(10)
y = (x-5)**2
print(x)
print(y)
```

```python
plt.style.use("seaborn")
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

#### GIven a function f(x) we want to find value of x that minimizes f.

```python
x=0
step_size = 0.1
error = []
#take 50 steps in downhil diretion
for i in range(30):
    gradient = 2*(x-5)
    x = x - (step_size*gradient)
    y = (x-5)**2
    error.append(y)
    print(x,y)
    plt.scatter(x,y)
plt.show()
plt.plot(error)
```
