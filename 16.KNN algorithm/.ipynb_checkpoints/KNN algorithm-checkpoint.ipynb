# KNN

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = pd.read_csv("../datasets/xdata.csv",header=None)
y = pd.read_csv("../datasets/ydata.csv",header=None)
```

```python
x=x.iloc[:,1:]
y=y.loc[:,1:]
```

```python
x = x.values
y = y.values
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
query_point=np.array([2,3])
plt.scatter(query_point[0],query_point[1],c='Red')
plt.show()
```

```python
import math
def Euc_distance(x,y,K=5):
    distance=[]
    for i in range(y.shape[0]):
        dis = math.sqrt((query_point[0]-x[i][0])**2+(query_point[1]-x[i][1])**2)
        distance.append((dis,y[i]))
    distance.sort()
    distance=distance[0:K]
    return distance
```

```python
Euc_distance(x,y)
```

```python
def KNN(x,y,K=5):
    distance=Euc_distance(x,y)
    classes=[]
    for i in range(K):
        classes.append(distance[i][1])
    return np.unique(classes,return_counts=True)[1]
    
    
```

```python
result=KNN(x,y)
np.argmax(result)
```
